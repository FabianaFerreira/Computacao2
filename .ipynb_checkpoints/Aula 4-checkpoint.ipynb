#### A seguir, vamos fazer a implementação de uma função que será usada na definição das funções `__init__` e `__setitem__` para checar se os valores são válidos (tipo `int` ou `float`)

```python
#Função auxiliar para uso no exercício 17
def checarValores(var, default=1):
    if (isinstance(var,int) or isinstance(var,float)):
        return var, True
    return default, False    
```

#### Redefinição da classe ponto3D para utilizar exceções

```python
class ponto3D:
    #Exercicio 15
    def __init__ (self,x,y,z):                        
        self.x, boolean = checarValores(x)
        if (not boolean):
            raise TypeError ("Valor inválido")
        self.y, boolean = checarValores(y)
        if (not boolean):
            raise TypeError ("Valor inválido")
        self.z, boolean = checarValores(z)
        if (not boolean):
            raise TypeError ("Valor inválido")
            
    #Exercicio 2
    def __str__ (self):
        return "(" + str(self.__x) + "," + str(self.__y) + "," + str(self.__z) + ")"

    #Exercicio 3
    def distancia (self, ponto):
        if (not isinstance(ponto, ponto3D)):
            return -1;

        x1 = self.__x
        y1 = self.__y
        z1 = self.__z
        x2 = ponto['x']
        y2 = ponto['y']
        z2 = ponto['z']
        
        return sqrt(pow(x1 - x2, 2) + pow(y1 - y2, 2) + pow(z1 - z2, 2))

    #Exercicio 4
    def __mod__(self, ponto):
        return self.distancia(ponto)

    #Exercicio 5
    def __getitem__ (self, key):
        if (key == 1 or key == 'x'):
            return self.__x
        if (key == 2 or key == 'y'):
            return self.__y
        if (key == 3 or key == 'z'):
            return self.__z
        return None
    
    #Exercicio 16
    def __setitem__ (self, key, value):        
        if (key == 1 or key == 'x'):            
            self.x, boolean = checarValores(value)
            if (not boolean):
                raise TypeError ("Valor inválido")
        if (key == 2 or key == 'y'):
            self.y, boolean = checarValores(value)
            if (not boolean):
                raise TypeError ("Valor inválido")
        if (key == 3 or key == 'z'):
            self.z, boolean = checarValores(value)
            if (not boolean):
                raise TypeError ("Valor inválido")
                
        if (self.dx == 0 and self.dy == 0 and self.dz == 0):
            raise ValueError ('Não é possível atualizar o vetor, pois criará um vetor degenerado de comprimento zero.')                    

    #Exercicio 12
    def __sub__ (self, ponto):
        return vector3D(ponto['x'] - self.__x,
                        ponto['y'] - self.__y,
                        ponto['z'] - self.__z)

    #Exercicio 13
    def __add__ (self, vector):
        if (not isinstance(vector, vector3D)):
            return -1
        return ponto3D(self.__x + vector['dx'], 
                       self.__y + vector['dy'],
                       self.__z + vector['dz'])
```

#### Redefinição da classe vector3D para utilizar exceções

```python
class vector3D:
    #Exercicio 15
    def __init__(self,dx,dy,dz):        
        self.__dx, boolean = checarValores(dx)
        if (not boolean):
            raise TypeError ("Valor inválido")
        self.__dy, boolean = checarValores(dy)
        if (not boolean):
            raise TypeError ("Valor inválido")
        self.__dz, boolean = checarValores(dz)
        if (not boolean):
            raise TypeError ("Valor inválido")
            
        if (dx == 0 and dy == 0 and dz == 0):
            raise ValueError ('Não é possível criar um vetor degenerado de comprimento zero.')

    #Exercicio 15
    def __getitem__ (self, key):
        if (key == 1 or key == 'dx'):
            return self.__dx
        if (key == 2 or key == 'dy'):
            return self.__dy
        if (key == 3 or key == 'dz'):
            return self.__dz
        return None
    
    #Exercicio 16
    def __setitem__ (self, key, value):
        if (key == 1 or key == 'dx'):            
            self.__dx, boolean = checarValores(value)
            if (not boolean):
                raise TypeError ("Valor inválido")
        if (key == 2 or key == 'dy'):
            self.__dy, boolean = checarValores(value)
            if (not boolean):
                raise TypeError ("Valor inválido")
        if (key == 3 or key == 'dz'):
            self.__dz, boolean = checarValores(value)
            if (not boolean):
                raise TypeError ("Valor inválido")
        
    #Exercicio 7
    def __abs__(self):
        return (self.__dx**2 + self.__dy**2 + self.__dz**2)**0.5

    #Extra para não ter que ficar convertendo com str()
    def __str__ (self):
        return "(" + str(self.__dx) + "," + str(self.__dy) + "," + str(self.__dz) + ")"
    
    #Exercicio 8
    def unitario(self):
        modulo = abs(self)
        xn = self.__dx/modulo
        yn = self.__dy/modulo
        zn = self.__dz/modulo
        return vector3D(xn,yn,zn)

    #Exercicio 9/14
    def __add__ (self, outro):
        if (isinstance(outro, ponto3D)):
            return outro + self
        return vector3D(self.__dx + outro['dx'], 
                        self.__dy + outro['dy'], 
                        self.__dz + outro['dz'])

    #Exercicio 10    
    def __neg__ (self):        
        return vector3D(self.__dx*(-1), self.__dy*(-1), self.__dz*(-1))

    #Exercicio 11    
    def __sub__ (self, v2): 
        return self + (-v2)    
    
    #Exercicio 18
    def produtoEscalar (self, vector):
        return float(self.__dx*vector['dx'] + self.__dy*vector['dy'] + self.__dz*vector['dz'])
```
